**Selecting the important features of the dataset and storing in a new dataframe**

**Data Visualization**

**1. Year Wise Attacks** 

```python
plt.figure(figsize=(25,7))
sns.countplot(x='iyear',data=df1,palette='Set1')
plt.title('Number of Attacks Every Year')
plt.xticks(rotation=45)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.show()
```

Maximum Number of Attack took place in the year 2014.


Minimum Number of Attack took place in the year 1971.

```python
yk=df1[['iyear','nkill']].groupby('iyear').sum()
yk.head()
```

```python

yk.min()
```

```python
yk.max()
```

```python
yk.plot(kind="bar",color='blue',figsize=(15,6))
plt.title("Year wise killed",fontsize=13)
plt.xlabel("Years",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of persons killed",fontsize=13)
plt.show()


```

**2. Region wise Attacks**

```python
reg=pd.crosstab(df1.iyear,df1.region_txt)
reg.head()
```

```python
reg.plot(kind="area",  alpha=0.5,figsize=(20,10))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

```python
regt=reg.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(15,6),color='red')
plt.title("Total Number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

Most number of attack took place in Middle East and North America along the years.

```python
rk=df1[['region_txt','nkill']].groupby('region_txt').sum()
rk.sort_values(by="nkill",ascending=False)
rk.head()
```

```python
rk.plot(kind='bar',figsize=(15,6))
plt.title('Region wise killed',fontsize=15)
plt.xlabel("Region",fontsize=20)
plt.ylabel("Killed",fontsize=20)
plt.show()
```

**3.Country wise attacks**


```python
ct=df1["country_txt"].value_counts().head(10)
ct
```

```python
ct.plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.title("Country wise Attacks",fontsize=13)
plt.xlabel("Countries",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Attacks",fontsize=13)
plt.show()
```

```python
cnk=df1[["country_txt","nkill"]].groupby("country_txt").sum()
cnk.sort_values(by="nkill",ascending=False)
cnk.head(10)
```

```python
cnk[:10].plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.title("People Killed in each Country")
plt.xlabel("Countries")
plt.ylabel("Number of People Killed")
```

**4. City wise attacks**

```python
lt=df1["city"].value_counts().head(10).drop('Unknown')
lt
```

```python
lt.plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.title("City wise Attacks",fontsize=13)
plt.xlabel("City",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Attacks",fontsize=13)
plt.show()
```

```python
ck=df[["city","nkill"]].groupby('city').sum().sort_values(by="nkill",ascending=False).drop("Unknown")
ck.head(10)
```

```python
ck[:10].plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.title("People Killed in each City")
plt.xlabel("Cities")
plt.ylabel("Number of People Killed")
```

Baghdad is the city with most kills.

**5. Attack wise**

```python
at=df1["attacktype1_txt"].value_counts()
at
```

```python
at.plot(kind="bar",figsize=(15,6))
plt.title("Types of Attacks",fontsize=13)
plt.xlabel("Attack Types",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Attacks",fontsize=13)
plt.show()
```

```python
ak=df1[["attacktype1_txt","nkill"]].groupby("attacktype1_txt").sum().sort_values(by="nkill",ascending=False)
ak
```

```python
ak.plot(kind="bar",figsize=(15,6))
plt.title("People Killed in each Attack Type")
plt.xlabel("Attack Types")
plt.ylabel("Number of people Killed")
plt.show()
```

**6. Weapon type**

```python
wt=df1["weaptype1_txt"].value_counts().head(5)
wt
```

```python
wt.plot(kind="bar",figsize=(15,6),color='yellow')
plt.title("Types of Weapons Used",fontsize=13)
plt.xlabel("Weapon Types",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Attacks",fontsize=13)
plt.show()
```
